Necessário para a leitura do pyAgrum

```python
variaveis_categoricas = ['asma_pre', 'cardiopatia_pre',
       'diabetes_pre', 'doenca_hematologica_pre', 'doenca_hepatica_pre',
       'doenca_neurologica_pre', 'doenca_renal_pre', 'imunodepressao_pre',
       'obesidade_pre', 'outros_fatores_de_risco_pre', 'pneumopatia_pre',
       'puerpera_pre', 'sindrome_de_down_pre','cs_sexo_pre']
```

```python
def label_x(df):
    return pd.concat(
           (pd.DataFrame(np.array([comor_encoder.transform(df[c]) for c in variaveis_categoricas 
                          if c != "cs_sexo_pre"]).T,
                        columns = variaveis_categoricas[:-1]),
            pd.Series(idade_encoder.fit_transform(pd.cut(df.idade_pre, bins = [-np.inf,27,36,45,57,np.inf])),
                     name = "idade_pre"),  #discretização do classificador 1
            pd.Series(genero_encoder.transform(df["cs_sexo_pre"]),name = "cs_sexo_pre"),
            df.obito_pre),
            axis = 1)
```

```python
#Exportando os dados
#vamos separar os dados em treino e teste, para avaliar se o nosso modelo possui capacidade de generalização
if (("comorbidade_pyAgrum_train.csv" not in os.listdir("../data/interim/")) | 
   ("comorbidade_pyAgrum_test.csv" not in os.listdir("../data/interim/"))):
    x_train_,x_test_,y_train_,y_test_ = train_test_split(df_comorbidades.drop(columns = ["municipio_FIOCRUZ",
                                                                           "nomeRegiaoSaude_FIOCRUZ",
                                                                           "obito_pre"]),
                                           df_comorbidades["obito_pre"],
                                           random_state = 21,
                                           stratify = df_comorbidades["obito_pre"])
    
    comor_encoder = LabelEncoder()
    genero_encoder = LabelEncoder()

    comor_encoder.fit(x_train_.asma_pre)
    genero_encoder.fit(x_train_.cs_sexo_pre)
    
    train = label_x(pd.concat((x_train_.reset_index(drop = True),y_train_.reset_index(drop = True)),axis=1))
    test = label_x(pd.concat((x_test_.reset_index(drop = True),y_test_.reset_index(drop = True)),axis=1))
    
    train.to_csv("../data/interim/comorbidade_pyAgrum_train.csv",index=False)
    test.to_csv("../data/interim/comorbidade_pyAgrum_test.csv",index=False)
    
    x_train,y_train = (train.drop(columns = "obito_pre"), train["obito_pre"])
    x_test, y_test = (test.drop(columns = "obito_pre"), test["obito_pre"])
    
else:
    comor_encoder = LabelEncoder()
    genero_encoder = LabelEncoder()
    comor_encoder.fit(df_comorbidades.asma_pre)
    genero_encoder.fit(df_comorbidades.cs_sexo_pre)
    
    train,test = (pd.read_csv("../data/interim/comorbidade_pyAgrum_train.csv"),
                  pd.read_csv("../data/interim/comorbidade_pyAgrum_test.csv"))
    
    x_train,y_train = (train.drop(columns = "obito_pre"), train["obito_pre"])
    x_test, y_test = (test.drop(columns = "obito_pre"), test["obito_pre"])
```

```python
learner = gum.BNLearner("../data/interim/comorbidade_pyAgrum_train.csv")
```

```python
#Vamos adicionar algumas conexões chave, ou remover as que não fazem sentido

"""A primeira está relacionada a que, dado uma comorbidade, é sabido que esta influencia na morte dos pacientes, 
mas as mortes dos pacientes não influencia diretamente"""
for vc in variaveis_categoricas:
    learner.addMandatoryArc(vc,"obito_pre") #Forçando que exista essa relação com o óbito
    learner.addForbiddenArc("obito_pre",vc) #o óbito deve ser um collider, mas deve ser o estado final
learner.addMandatoryArc("idade_pre","obito_pre")

"""Nenhuma comorbidade gera a sindrome de Down, então, não vamos deixar acontecer. O mesmo vale para puerpera, idade,
genero"""
independentes = ["sindrome_de_down_pre","idade_pre","puerpera_pre","cs_sexo_pre"]
for i in independentes:
    for c in [v for v in variaveis_categoricas if i != v]:
        learner.addForbiddenArc(c,i)

"""A idade não influencia a presença de sindrome de down ou gênero do individuo"""
learner.addForbiddenArc("idade_pre","cs_sexo_pre")
learner.addForbiddenArc("idade_pre","sindrome_de_down_pre")

"""Somente mulheres pondem ficar puerperas, então..."""
learner.addMandatoryArc("cs_sexo_pre","puerpera_pre")
```

# Construção do primeiro classificador

Vamos fazer a construção de um classificador e comparar seus resultados com a primeira rede Baysiana criada

```python
def plot_confusion_matrix(x,y,classifier):
    y_true = y
    y_pred = classifier.predict(x)
    y_score = classifier.predict_proba(x)[:,1]
    
    cm = confusion_matrix(y_true=y_true,y_pred=y_pred)
    print("Acc:",accuracy_score(y_true,y_pred))
    print("Balanced Acc:",balanced_accuracy_score(y_true,y_pred))
    print("F1 score:",f1_score(y_true,y_pred))
    print("Precision:",precision_score(y_true,y_pred))
    print("Recall:",recall_score(y_true,y_pred))
    print("AUC:",roc_auc_score(y_true,y_score))

    sns.heatmap(cm,annot=True,cbar = False)
    plt.xlabel("Predição")
    plt.ylabel("Real")
```

```python
def arestas_names_BN(BN):
    index_names_bn = {bn2.idFromName(n): n for n in bn2.names()}
    arestas_bn = [(index_names_bn[v[0]], index_names_bn[v[1]]) for v in bn2.arcs()]
    return arestas_bn
```

```python
#Criando o template
def cria_template_fromBN(BN,defaultDiscretizationMethod = "uniform",defaultNumberOfBins=10):
    learn = gum.BNLearner("../data/interim/comorbidade_pyAgrum_train.csv")
    for are in arestas_names_BN(BN):
        learn.addMandatoryArc(are[0],are[1])
    return learn
```

Podemos ajustar algumas hiperparâmetros, para a geração do nosso modelo (em negrito, o padrão):
* método de aprendizado: Chow-Liu, NaiveBayes, TAN (Tree
Augmented Naive Bayes), MIIC, __GHC__, Tabu ...;
* suavização a priori: Smoothing, BDeu, Dirichlet, NoPrior, __None__
* métrica: AIC, __BIC__, BDeu, K2, Log2AIC
* método de discretização: __quantile__, uniform, kmeans, NML (Normalized Maximum Likelihood), CAIM (Class-Attribute Interdependence Maximization), MDLP (Minimum Description Length Procedure)

Vamos, primeiramente, analisar as mudanças separadamente, e depois fazer algumas combinações

```python
def testa_modelo_diferentes_hiperparametros(kargs):
    results = {}
    combinacoes_parametros = list(product(*list(kargs.values())))
    for hp in combinacoes_parametros:
        kargs_aux = {k:v for k,v in zip(kargs.keys(),hp)}
        print(kargs_aux)
        clas = skbn.BNClassifier(**kargs_aux)
        clas.fit(x_train,y_train)
        results[hp] = balanced_accuracy_score(y_test,clas.predict(x_test))
    return results
```

```python
metodos_aprendizado = ["Chow-Liu", "NaiveBayes","TAN","GHC","Tabu","MIIC"]
resultado_metodo_aprendizado = testa_modelo_diferentes_hiperparametros({"learningMethod":metodos_aprendizado})
```

## Teste com ciclo 1, 2 e 3

### Ciclo 1

<b>Ideia</b>: realizar uma busca para encontrar as possíveis conexões que possam existir, ou seja, estamos em busca de uma combinação ótima de arestas. Por meio de buscas variando as formas de métricas, algoritmos de busca, métodos aPrior, aprende-se n redes diferentes e verifica as arestas que mais aparecem
    

```python
score_metrics = [ 'useScoreAIC', 
                 'useScoreBDeu', 
                 'useScoreBIC', 
                 'useScoreK2', 
                 'useScoreLog2Likelihood' ] # Bayesian Dirichlet equivalent uniform - BDeu

#'useK2' ***method removed temporarily while it isn't clear how to use the order list required by it***
search_algs = ['use3off2', 
               'useEM',
               'useGreedyHillClimbing',
               'useLocalSearchWithTabuList' ] 

aPrioriMethod = ["useAprioriSmoothing","useAprioriBDeu"]
aPrioriWeight = [1,1]

"""positive const > 0 for using EM, if epsilon = 0.0, then EM isn't used -- # 
epsilon was varied, no impact on essential graphs from EM was found
Next, order list for useK2 is empty since it wasn´t required when generating only essencial graphs useK2
order_list4K2 = [ 0 ] # apriori - domain prior knowledge from an expert # useK2 requires no empy list to work. 
                      I chose a single node list, since it in fact isn't an order, 
                      then it must not make any influence on BN constructions  """
epsilon = 0.1

order_list4K2 = [ ]            
```

```python
learner = gum.BNLearner("../data/interim/comorbidade_pyAgrum_train.csv")

bn0=learner.learnBN() #save a BN as reference of number of nodes and other features used by other pieces of code

"""maxindegree corresponds to the maximum of the ceil of the half of the degree of a node from a undirected graph 
with n nodes bn0.size() / 2"""

maxindegree = int(np.ceil(bn0.size()/2))
```

```python
def bns_all_metrics_searchalgorithms(learner, 
                                     maxindegree=1, 
                                     score_metrics=['useScoreAIC'], 
                                     search_algs=['useGreedyHillClimbing'],
                                     aPrioriMethod = [None],
                                     aPrioriWeight = [1],
                                     candidate_connections=[ ], 
                                     howtouse_candidates='addPossibleEdge', 
                                     apriori=[ ], 
                                     verbose=False): #learner - imported BN with variable set and cpt
    #Solving conflicts of a same edge as mandatory and forbidden
    #- mandatory is learned by the algorithm - it is data and method sensitivy
    #- forbidden is setup by the user
    #- then, fobidden edge wins
    print()
    print(candidate_connections)
    print(apriori)
    if howtouse_candidates!='setPossibleSkeleton': 
        candidate_connections = set(candidate_connections) #inicializa os possíveis candidatos
        apriori = set(apriori) #inicializa apriori
        candidate_connections = candidate_connections - apriori #fica apenas com as arestas únicas

    acc_edges = []
    acc_arcs  = []
    bns       = []
    for alg in search_algs:
        if verbose:
            print("\n\n Using Search Algorithm: {}".format(alg))
        f_alg = getattr(learner, alg)
        for scor in score_metrics:
            #Indegree setup
            learner.setMaxIndegree(maxindegree) #não permite que a rede tenha, para cada nó, mais que maxInDegree pais

            #score and search algorithms choices
            f_score = getattr(learner, scor)
            f_score() #e.g: learner.useScoreK2()
            if verbose:
                print("\nUsing score metrics: {}".format(scor))
            if alg == 'useEM':
                f_alg(epsilon)
            if alg == 'useK2':
                f_alg(order_list4K2)
            if (alg!='useEM') & (alg!='useK2'):
                f_alg() #e.g: learner.useLocalSearchWithTabuList()
                
            for prior,w in zip(aPrioriMethod,aPrioriWeight):
                f_aPriori = getattr(learner, prior)
                if verbose:
                    print("\nUsing aPriori: {}".format(prior))
                if prior == "useAprioriSmoothing":
                    learner.useAprioriSmoothing(w)
                if prior == "useAprioriBDeu":
                    learner.useAprioriBDeu(w)
                if (prior != "useAprioriBDeu") & (prior != "useAprioriSmoothing"):
                    f_aPriori()
                    
                #ways to use candidate connections
                ge = {} # in case no learn proceeds - howtouse_candidates do nothing
                if howtouse_candidates == 'addPossibleEdge': 
                    [ learner.addPossibleEdge(u, v) for u, v in candidate_connections ]
                    bn=learner.learnBN()
                    ge=gum.EssentialGraph(bn) #o grafo essencial é aquele mantem todas as propriedades de Markov
                                              #ou seja, um grafo que, independente de suas conexoes, irá manter 
                                              #sempre o mesmo esqueleto e as imoralidades
                    acc_edges += [ ge.edges() ]
                elif howtouse_candidates == 'setPossibleSkeleton':
                    learner.setPossibleSkeleton(skeleton=candidate_connections)
                    bn=learner.learnBN()
                    ge=gum.EssentialGraph(bn)
                    acc_edges += [ ge.edges() ]
                else:
                  #Mandatory and Fobidden work together BUT NEVER together with 
                  #addPossibleEdge or setPossibleSkeleton, even for empty sets of candidate connections
                  #howtouse_edges != 'Previous Options':  =='addMandatoryArc' or anything else - default case
                    [ learner.addMandatoryArc(u, v) for u, v in candidate_connections ]
                    [ learner.addForbiddenArc(u, v) for u, v in apriori ]
                    bn=learner.learnBN()
                acc_arcs  += [ bn.arcs()  ]
                bns       += [ bn         ] 

                if verbose:
                    ge
                    gnb.sideBySide(bn,ge)

    return acc_edges, acc_arcs, bns #retorna todas as arestas (não direcionadas), arcos (direcionado) e as BN
```

```python
def select_essential_edges(acc_edges, percentage=0.5):
    edge_count_dic = {}
    for elst in acc_edges: #para cada conjunto de arestas das N BN...
        for e in elst: #para cada aresta ...
            if e in edge_count_dic.keys(): 
                edge_count_dic[e] += 1
            else: 
                edge_count_dic[e]  = 1 #conta a quantidade de vezes que a aresta apareceu no processo "bns_all_metrics_searchalgorithms"
    #ordena pela quantidade de vezes que aparece e pega os "percentage" primeiros
    edge_dic_sorted = sorted(edge_count_dic.items(), key = lambda kv:(kv[1], kv[0]), reverse=True)
    selected_edges = [ k for k, v in edge_dic_sorted[0:int(len(edge_dic_sorted)*percentage+1)] ]
    return selected_edges, edge_dic_sorted

def cycle_1(learner, 
            maxindegree=1, 
            score_metrics=score_metrics, 
            search_algs=search_algs, 
            aPrioriMethod = aPrioriMethod,
            aPrioriWeight = aPrioriWeight):
    #Edge selected based on ranking essential edges 3off2
    possible_edges = [] # no essential edge found before starting the investigation

    acc_edges, acc_arcs, bns = bns_all_metrics_searchalgorithms(learner, 
                                                                maxindegree=maxindegree, 
                                                                score_metrics=score_metrics, 
                                                                search_algs=search_algs,
                                                                aPrioriMethod = aPrioriMethod,
                                                                aPrioriWeight = aPrioriWeight,
                                                                candidate_connections=possible_edges,
                                                                howtouse_candidates='addPossibleEdge',
                                                                verbose=False)
    possible_edges, edge_dic_sorted = select_essential_edges(acc_edges, percentage=1.0)

    # possible_edges = essential edges obtained by using 3off2 'alone' - the most frequent - 
    # usually the best half - percentage = 0.5
    possible_edges_0 = possible_edges 
    converged = False
    while not converged: #enquanto não convergir...
        #agora em mãos das possíveis conexões, vamos rodar o algoritmo de busca mais uma vez, auxiliando com as
        #possíveis conexões
        acc_edges, acc_arcs, bns = bns_all_metrics_searchalgorithms(learner, 
                                                                    maxindegree=maxindegree, 
                                                                    score_metrics=score_metrics, 
                                                                    search_algs=search_algs,
                                                                    aPrioriMethod = aPrioriMethod,
                                                                    aPrioriWeight = aPrioriWeight,
                                                                    candidate_connections=possible_edges,
                                                                    howtouse_candidates='addPossibleEdge',
                                                                    verbose=False)
        possible_edges, edge_dic_sorted = select_essential_edges(acc_edges, percentage=0.5) 
        #Verifica o quão diferente foram as possíveis arestas do processo atual com o processo anterior
        diff = set(possible_edges_0) - set(possible_edges) 
        if len(diff) != 0:
            converged = False
            possible_edges_0 = possible_edges
        else:
            converged = True
    return possible_edges
```

### Ciclo 2

__Ideia__: baseado nas informações obtidas do ciclo anterior, aprender a construir uma rede baysiana apenas com as aretas necessárias. 

```python
def make_a_skeleton(possible_edges, verbose=False):
    #Criação de um esqueleto, baseado nas possíveis combinações que podem existir
    skeleton = gum.UndiGraph()
    skeleton.addNodes(bn0.size())
    [ skeleton.addEdge(e[0],e[1]) for e in possible_edges ]
    if verbose:
        #gnb.showGraph(skeleton)
        print(skeleton)
    return skeleton

def cycle_2(learner, possible_edges, 
            maxindegree=1, 
            score_metrics=score_metrics, 
            search_algs=search_algs,
            aPrioriMethod = aPrioriMethod,
            aPrioriWeight = aPrioriWeight):
    possible_edges_0 = possible_edges #essential edges obtained by using 3off2 'alone'
    converged = False
    #enquanto não houver convergência
    while not converged:
        #cria o esqueleto
        skeleton = make_a_skeleton(possible_edges, verbose=False)
        #gera-se os possíveis arcos e vértices, baseado no esqueleto gerado
        acc_edges, acc_arcs, bns = bns_all_metrics_searchalgorithms(learner, 
                                                                    score_metrics=score_metrics, 
                                                                    search_algs=search_algs, 
                                                                    candidate_connections=skeleton, 
                                                                    howtouse_candidates='setPossibleSkeleton', 
                                                                    verbose=False,
                                                                    aPrioriMethod = aPrioriMethod,
                                                                    aPrioriWeight = aPrioriWeight)
        #Seleciona os vértices mais relevantes
        possible_edges, edge_dic_sorted = select_essential_edges(acc_edges, percentage=0.5)
        diff = set(possible_edges_0) - set(possible_edges)
        if len(diff) != 0:
            converged = False
            possible_edges_0 = possible_edges
        else:
            converged = True
    return possible_edges, acc_arcs
```

### Ciclo 3

<b>Ideia</b>: com o retorno do ciclo 2, que são os arcos (as arestas direcionadas mais frequentes), vamos criar um novo modelo, forçando a existência dessas arestas

```python
def sort_arcs(acc_arcs, reverse=True):
    arc_dic = {}
    for arclst in acc_arcs:
        for arc in arclst:
            if arc in arc_dic.keys(): 
                arc_dic[arc] += 1
            else:
                arc_dic[arc]  = 1
    arc_dic_sorted = sorted(arc_dic.items(), key = lambda kv:(kv[1], kv[0]), reverse=reverse)
    return arc_dic_sorted

#Keep one from arc (i,j) and arc (j,i), the most frequent one
def select_essential_corresponding_arcs(acc_arcs, verbose=True):
    arc_dic_sorted = sort_arcs(acc_arcs, reverse=False)
    if verbose:
        print(arc_dic_sorted)
    setkeep = set()
    [ setkeep.add(k) for k, v in arc_dic_sorted ]
    for k, v in arc_dic_sorted:
        er = (k[1],k[0])
        if er in setkeep:
            setkeep.remove(k)
    return setkeep

def set_orderlist4useK2(mandatory_arcs):
    nodeset = set()
    for a in mandatory_arcs:
        nodeset.add(a[0]), nodeset.add(a[1])
    order_list4K2 = [ n  for n in nodeset ]
    return order_list4K2

def cycle_3(learner, 
            acc_arcs, 
            maxindegree=1, 
            score_metrics=score_metrics, 
            search_algs=search_algs, 
            apriori=[ ], 
            verbose=True,
            aPrioriMethod = aPrioriMethod,
            aPrioriWeight = aPrioriWeight):
    #os arcos mais frequentes, provenientes do ciclo 2
    mandatory_arcs = select_essential_corresponding_arcs(acc_arcs, verbose=True)
    #if search algorithm useK2 is used, 'a node order list' is one of its arguments
    order_list4K2 = set_orderlist4useK2(mandatory_arcs)
    acc_edges, acc_arcs, bns = bns_all_metrics_searchalgorithms(learner, 
                                                                score_metrics=score_metrics, 
                                                                search_algs=search_algs, 
                                                                candidate_connections=mandatory_arcs, 
                                                                howtouse_candidates='addMandatoryArc', 
                                                                apriori=apriori, 
                                                                verbose=verbose,
                                                                aPrioriMethod = aPrioriMethod,
                                                                aPrioriWeight = aPrioriWeight)
    return bns
```

### Comparação par a par das redes

```python
def dict2html(di1,di2=None,di3=None): 
    res= "<br/>".join([f"<b>{k:15}</b>:{v}" for k,v in di1.items()])
    if di2 is not None:
        res+="<br/><br/>"
        res+= "<br/>".join([f"<b>{k:15}</b>:{v}" for k,v in di2.items()])
    if di3 is not None:
        res+="<br/><br/>"
        res+= "<br/>".join([f"<b>{k:15}</b>:{v}" for k,v in di3.items()])
    return res

def comp_2bns(bn1, bn2, verbose=False):
    kl=gum.ExactBNdistance(bn1,bn2) # bruteForce is possible since the BNs are small
    cmp = gcm.GraphicalBNComparator(bn1,bn2)
    if verbose:
        gnb.sideBySide(bn1,bn2,gnb.getBNDiff(bn1,bn2),dict2html(cmp.scores(),cmp.hamming()), 
                       cmp.skeletonScores(),  
                       cmp.equivalentBNs(),
                       dict2html(kl.compute()),
                       captions=['bn1','bn2','Graphical diffs','Scores', 'Skeleton scores', 'Equivalent ?','Distances'])
    scores = cmp.scores()
    dsts_kl = kl.compute()
    return dsts_kl['klPQ'], dsts_kl['klQP'], dsts_kl['hellinger'], dsts_kl['jensen-shannon'], scores['dist2opt']
```

```python
class fitness_tuple:
    def __init__(self, a=0, b=0):
        self.fitness = (a,b)

    def __iter__(self):
        return self.fitness

# Define the key field for sorting
def get_key(obj):
    return obj.fitness

# Sort function: fast non-dominated sort
def nsgaii(population, maximise):
    # Get population size
    popsize = len(population)
    poprange = range(popsize)
    #
    # Initialize the sorted population
    front = [[]]
    # Initialize the set of solutions that are dominated
    s = [[] for i in poprange]
    # Initialize the number of solutions which dominate an individual
    n = [0 for i in poprange]
    # Initialize the rank of solutions
    rank = [0 for i in poprange]
    #
    # Looks for the `p`-dominated solutions and
    # calculates the degree of domination over `p`
    for p in poprange:
        s[p] = []
        n[p] = 0
        # Get `p` fitness
        pfa, pfb = population[p].fitness

        #print(p, pfa, pfb)

        for q in poprange:
          # Get `q` fitness
            qfa, qfb = population[q].fitness
          #
          # Maximise `p` domination
            if maximise:
            # Check which dominates which
                if ((pfa > qfa and pfb > qfb) or (pfa >= qfa and pfb > qfb)
                    or (pfa > qfa and pfb >= qfb)):
                        s[p].append(q)
                elif ((qfa > pfa and qfb > pfb) or (qfa >= pfa and qfb > pfb)
                      or (qfa > pfa and qfb >= pfb)):
                        n[p] += 1
              #
              # Minimise `p` domination
            else:
            # Check which dominates which
                if ((pfa < qfa and pfb < qfb) or (pfa <= qfa and pfb < qfb)
                  or (pfa < qfa and pfb <= qfb)):
                    s[p].append(q)
                elif ((qfa < pfa and qfb < pfb) or (qfa <= pfa and qfb < pfb)
                  or (qfa < pfa and qfb <= pfb)):
                    n[p] += 1
        #
        # Check if `p` belongs to the fisrt front
        if n[p] == 0:
            rank[p] = 0
            if p not in front[0]:
                front[0].append(p)
    #
    # Initiliaze the front counter
    i = 0
    while front[i] != []:
        aux = []
        for p in front[i]:
            for q in s[p]:
                n[q] = n[q] - 1
                if n[q] == 0:
                    rank[q] = i + 1
                    if q not in aux:
                        aux.append(q)
        i += 1
        front.append(aux)
  # Remove the last set of individuals
    del front[len(front) - 1]
    #  #
    #  # Convert to a usual population list
    #  sorted_pop = []
    #  for f in front:
    #    for i in f:
    #      sorted_pop.append(population[i])
    #  return sorted_pop
    return front

# Select function: fast non-dominated selection
def nsgaii_select(population, maximise):
    # Get population size
    popsize = len(population)
    poprange = range(popsize)
    #
    # Initialize the sorted population
    front = [[]]
    # Initialize the set of solutions that dominates # are dominated
    s = [[] for i in poprange]
    # Initialize the number of solutions which dominate an individual
    n = [0 for i in poprange]
    # Initialize the rank of solutions
    rank = [0 for i in poprange]
    #
    # Looks for the `p`-dominated solutions and
    # calculates the degree of domination over `p`
    for p in poprange:
        s[p] = []
        n[p] = 0
        # Get `p` fitness
        pfa, pfb = population[p].fitness
        for q in poprange:
          # Get `q` fitness
            qfa, qfb = population[q].fitness
            #
            # Maximise `p` domination
            if maximise:
                # Check which dominates which
                if ((pfa > qfa and pfb > qfb) or (pfa >= qfa and pfb > qfb)
                  or (pfa > qfa and pfb >= qfb)):
                    s[p].append(q)
                elif ((qfa > pfa and qfb > pfb) or (qfa >= pfa and qfb > pfb)
                  or (qfa > pfa and qfb >= pfb)):
                    n[p] += 1
            #
            # Minimise `p` domination
            else:
            # Check which dominates which
                if ((pfa < qfa and pfb < qfb) or (pfa <= qfa and pfb < qfb)
                  or (pfa < qfa and pfb <= qfb)):
                    s[p].append(q)
                elif ((qfa < pfa and qfb < pfb) or (qfa <= pfa and qfb < pfb)
                  or (qfa < pfa and qfb <= pfb)):
                    n[p] += 1
        #
        # Check if `p` belongs to the fisrt front
        if n[p] == 0:
            rank[p] = 0
            if p not in front[0]:
                front[0].append(p)
  #
  # Initiliaze the front counter
    i = 0
    while front[i] != []:
        aux = []
        for p in front[i]:
            for q in s[p]:
                n[q] = n[q] - 1
                if n[q] == 0:
                    rank[q] = i + 1
                    if q not in aux:
                        aux.append(q)
        i += 1
        front.append(aux)
  # Remove the last set of individuals
    del front[len(front) - 1]
    return front[0]

def biobjetive_frontiers(serie_obj1, serie_obj2, maximise=True):
    pair = zip(serie_obj1, serie_obj2)
    population = []
    for p,q in pair:
        population.append( fitness_tuple(p,q) )
    sorted_pop = nsgaii(population, maximise=maximise)
    #sorted_pop
    np_spop = np.array(sorted_pop) # reformating it for improving presentation
    return np_spop
```

```python
def dict2html(di1,di2=None,di3=None): 
    res= "<br/>".join([f"<b>{k:15}</b>:{v}" for k,v in di1.items()])
    if di2 is not None:
        res+="<br/><br/>"
        res+= "<br/>".join([f"<b>{k:15}</b>:{v}" for k,v in di2.items()])
    if di3 is not None:
        res+="<br/><br/>"
        res+= "<br/>".join([f"<b>{k:15}</b>:{v}" for k,v in di3.items()])
    return res

def comp_2bns(bn1, bn2, verbose=False):
    kl=gum.ExactBNdistance(bn1,bn2) # bruteForce is possible car the BNs are small
    cmp = gcm.GraphicalBNComparator(bn1,bn2)
    if verbose:
        gnb.sideBySide(bn1,bn2,gnb.getBNDiff(bn1,bn2),dict2html(cmp.scores(),cmp.hamming()), 
                       cmp.skeletonScores(),  
                       cmp.equivalentBNs(),
                       dict2html(kl.compute()),
                       captions=['bn1','bn2','Graphical diffs','Scores', 'Skeleton scores', 'Equivalent ?','Distances'])
    scores = cmp.scores()
    dsts_kl = kl.compute()
    return (dsts_kl['klPQ'], 
           dsts_kl['klQP'], 
           dsts_kl['hellinger'], dsts_kl['jensen-shannon'], scores['dist2opt'], dsts_kl['bhattacharya'], scores['fscore'])
```

```python
#find the simplest BN
maxdim = 99999
for bn in bns_feature_dic.values():
    ndim = bn.dim()
    #print(ndim,bn), gnb.showBN(bn)
    if maxdim > ndim:
        maxdim = ndim
        bn_simplest = bn
```

```python
pairsPQ_QP = []
label = 1
for bn in bns_feature_dic.values():
    klPQ, klQP, hellinger, jensen_shannon, dist2plot, bhattacharya, fscore = comp_2bns(bn_simplest, bn, verbose=True)
    pairsPQ_QP  += [ ( label, klPQ, klQP, hellinger, jensen_shannon, bhattacharya, dist2plot, fscore, bn_simplest, bn, 0 ) ]
    label += 1

dfPQ_QP2simplestBN = pd.DataFrame( pairsPQ_QP, columns = ['Pair Label', 'klPQ', 'klQP', 'hellinger', 'jensen-shannon', 
                                                          'bhattacharya', 'dist2plot', 'fscore', 'bn_simplest', 'bn', 'Fronteira'] )
```
